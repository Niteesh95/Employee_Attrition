The columns 'Over18' and 'StandardHours' can be dropped from the data because all the records have the same value and it does not contain any variance. Therefore, We cannot infer much information from it about the employee attrition.

```python
#drop over 18 and sstd hours here
df.drop(['Over18','StandardHours'], inplace=True, axis=1)
```

```python
total = len(df_left['Gender'])
def my_fmt(x):
    # print(x)
    return '{:.1f}%\n({:.0f})'.format(x, total*x/100)

f,ax = plt.subplots(1,2, figsize=(18,6))
df_left['Gender'].value_counts().plot.pie(autopct=my_fmt,shadow=True, ax=ax[0]) 

sns.countplot(data=df, x='Gender', hue='Left', ax=ax[1])
```

```python
sns.countplot(data=df_left, x='MonthlyIncome', hue='PerformanceRating', order=df_left['MonthlyIncome'].value_counts().index)
plt.legend(bbox_to_anchor=(1,0.85), title="Performance Rating", fontsize=7, loc='right')
plt.xlabel('Monthly Income of Employees who left', fontsize=14)
plt.grid(visible=True, which='both')
plt.show()
```

```python
sns.countplot(data=df_stayed, x='MonthlyIncome', hue='PerformanceRating')
plt.legend(bbox_to_anchor=(1,0.85), title="Performance Rating", fontsize=7, loc='right')
plt.xlabel('Monthly Income of Employees who stayed', fontsize=14)
plt.grid(visible=True, which='both')
plt.show()
```

```python
sns.countplot(data=df_left, x='MonthlyIncome', hue='JobSatisfaction')
plt.legend(bbox_to_anchor=(1,0.85), title="Job Satisfaction", fontsize=7, loc='right')
plt.xlabel('Monthly Income', fontsize=14)
plt.grid(visible=True, which='both')
plt.show()
```

```python
sns.countplot(data=df, x='AgeBins', hue='Left')
plt.xlabel('Age of employees in years')
plt.legend(labels=['stayed', 'left'])
```

```python
#Feature Engineer Performance - 1,2,3 -> Low and 4,5 -> High
df["PerformanceRating_bool"] = df['PerformanceRating'].apply(lambda x:'High' if x>3 else "low")
```

```python
performance = df.loc[df['PerformanceRating_bool']=='High']
# compResN = comp.loc[comp['MonthlyIncome']=='N']
print(performance.groupby(['MonthlyIncome', 'Left'])['Left'].count())
sns.countplot(data=performance, x='MonthlyIncome', hue='Left', order=['low','medium','high'])
plt.xlabel('Monthly income of employees with high performance rating')
plt.legend(labels=['stayed', 'left'])
```

The reason for the high performing employees' attrition is they are not satisified with the income they get, they feel they are paid less and hence tend to switch, which often gives high increment for their current salary.

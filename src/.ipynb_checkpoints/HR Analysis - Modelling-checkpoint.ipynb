## Data Cleaning and Feature Engineering

```python
df.info()
```

```python
# Validate null values
for col in df.columns:
    if df[col].isna().sum() > 0:
        print('{} has {} null values'.format(col, df[col].isna().sum()))
```

```python
#drop over 18 and sstd hours here
df.drop(['Over18','StandardHours'], inplace=True, axis=1)
```

```python
#lets create age groups
df['AgeBins'] = pd.cut(df['Age'],bins=[9,19,29,39,49,59,69],labels=["10-19","20-29","30-39","40-49","50-59", "60-69"])
```

```python
df['complaintresolved'].fillna('NoComplaintFiled', inplace=True)
df['complaintyears'].fillna(0, inplace=True)
# sns.countplot(data=df, x='', hue='Left')
```

```python
Left = {'Yes': 1,'No': 0}
  
df['Left'] = [Left[item] for item in df['Left']]
df['Left'].value_counts()
```

```python
#Feature Engineer Performance - 1,2,3 -> Low and 4,5 -> High
df["PerformanceRating_bool"] = df['PerformanceRating'].apply(lambda x:'High' if x>3 else "low")
```

```python
df['%SalHike_Bins'] = pd.cut(df['PercentSalaryHike'],bins=[10,15,20,25],labels=["11-15%","16-20%","21-25%"])
```

```python
df.head()
```

```python

correlations = df.corr()
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(correlations, annot=True)
```

```python
# !pip install category_encoders
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordEncoder = OrdinalEncoder(categories=[['low', 'medium', 'high']])
df[['income_encoded']]=ordEncoder.fit_transform(df[['MonthlyIncome']])
ordEncoder.categories_
```

```python
income = {'low': 1,'medium': 2, 'high':3}
  
df['income_encoded'] = [income[item] for item in df['MonthlyIncome']]
df['income_encoded'].value_counts()
```

```python
cat_dummies = pd.get_dummies(df[['Gender','Department', 'BusinessTravel']])
```

```python
df['complaintresolved'].value_counts()
```

```python
complaint = {'Y': 1,'NoComplaintFiled': 1, 'N':0}
  
df['complaintResolved_encoded'] = [complaint[item] for item in df['complaintresolved']]
df['complaintResolved_encoded'].value_counts()
```

## ML Models - without oversampling (SMOTE)

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

inputs = [col for col in df_train1.columns if col != 'Left']
X = df_train1[inputs]
y = df_train1[['Left']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=50, stratify=y)
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

### Decision Tree model - without SMOTE

```python
#GridSearchCV to find best parameters for decision tree model
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

grid_params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'max_depth' : [3, 5, 7, 10],
    # 'min_samples_split' : range(2, 10, 1),
    # 'min_samples_leaf' : range(2, 10, 1),
    'max_features' : range(2,11,1),
    'random_state': range(0,50,5)
} 
   
grid = GridSearchCV(DecisionTreeClassifier(), grid_params, refit = True, verbose = 3,n_jobs=-1, scoring='accuracy', cv=5) 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 
# print best parameter after tuning 
print(grid.best_params_) 
grid_predictions = grid.predict(X_test)
```

```python
dtc = DecisionTreeClassifier(criterion='gini', max_depth=5, max_features=6, random_state=15, class_weight={0: 1, 1: 1})
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy : {dtc_train_acc}")
print(f"Test accuracy : {dtc_test_acc}")

#kfold cross vallidation
from sklearn.model_selection import cross_val_score, StratifiedKFold

kfold = StratifiedKFold(n_splits=5)
cv = cross_val_score(dtc,X_train,y_train,cv=kfold)
print(np.mean(cv))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = dtc.classes_)
disp.plot(cmap=plt.cm.Reds)
plt.title("Decision Tree Confusion Matrix")
plt.grid(visible=False)
plt.show()

print(classification_report(y_test, y_pred))
```

```python
df_train1.head(2)
```

```python
y_pred = dtc.predict(X_test)
AUC_dtc = dtc.predict_proba(X_test)
ROC_AUC_Score_dtc = round(metrics.roc_auc_score(y_test,AUC_dtc[:,1]), 3)
fpr_dtc, tpr_dtc, _ = metrics.roc_curve(y_test, y_pred)
```

### Random Forest Model - without SMOTE

```python
#GridSearchCV to find best parameters for random forest model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

params = {'criterion' : ['gini', "entropy", "log_loss"], 
          'max_depth' : range(2,11,1),
         'n_estimators' : range(2,11,1),
         'n_jobs': [2,3,4,5]}

grid = GridSearchCV(RandomForestClassifier(), params, refit = True, verbose = 3,n_jobs=-1, scoring='accuracy', cv=5) 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 
# print best parameter after tuning 
print(grid.best_params_) 
grid_predictions = grid.predict(X_test)
```

```python
#Random forest model

rfc = RandomForestClassifier(criterion='gini', max_depth=5,  n_estimators=6, n_jobs=3)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

rfc_train_acc = accuracy_score(y_train, rfc.predict(X_train))
rfc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy : {rfc_train_acc}")
print(f"Test accuracy : {rfc_test_acc}")

#kfold cross vallidation
from sklearn.model_selection import cross_val_score, StratifiedKFold

kfold = StratifiedKFold(n_splits=5)
cv = cross_val_score(rfc,X_train,y_train,cv=kfold)
print(np.mean(cv))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = rfc.classes_)
disp.plot(cmap=plt.cm.Reds)
plt.title("Random Forest Confusion Matrix")
plt.grid(visible=False)
plt.show()

print(classification_report(y_test, y_pred))
```

```python
#feature importances
Features_important={}
featureNames = X.columns.values
important = rfc.feature_importances_
for i in range(len(important)):
    Features_important[featureNames[i]]=important[i]

Important_features = dict(sorted(Features_important.items(),key= lambda x:x[1],reverse=True))
```

## SMOTE

```python
# !pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE

inputs = [col for col in df_train1.columns if col != 'Left']
X = df_train1[inputs]
y = df_train1[['Left']]

smote = SMOTE()
X,y = smote.fit_resample(X,y)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,stratify=y,random_state=50)
```

### Decision Tree model - with SMOTE

```python
dtc = DecisionTreeClassifier(criterion='gini', max_depth=6, max_features=5, random_state=15)
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, y_pred)

print(f"Training accuracy : {dtc_train_acc}")
print(f"Test accuracy : {dtc_test_acc}")

#kfold cross vallidation
from sklearn.model_selection import cross_val_score, StratifiedKFold

kfold = StratifiedKFold(n_splits=5)
cv = cross_val_score(dtc,X_train,y_train,cv=kfold)
print(np.mean(cv))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = dtc.classes_)
disp.plot(cmap=plt.cm.Reds)
plt.title("Decision Tree Confusion Matrix")
plt.grid(visible=False)
plt.show()

print(classification_report(y_test, y_pred))
```
